```python
conn = psycopg2.connect("dbname=wrangling user=postgres password=postgres")

# Open a cursor to perform database operations
cur = conn.cursor()


conn.autocommit = True
# Execute a query
cur.execute("Create database new_db;")

# Retrieve query results
cur.close()
conn.close()
```

```python
conn = psycopg2.connect("dbname=new_db user=postgres password=postgres")

# Open a cursor to perform database operations
cur = conn.cursor()


conn.autocommit = True
# Execute a query
cur.execute("""Create table new_table;
                id serial primary key,
                name varchar""")

# Retrieve query results
cur.close()
conn.close()
```
